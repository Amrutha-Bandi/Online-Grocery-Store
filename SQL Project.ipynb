```python
# 3. Total sales by product category
product_sales = orderdetails_df.merge(products_df, on='ProductID')
category_sales = product_sales.groupby('Category')['TotalPrice'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='TotalPrice', y='Category', data=category_sales)
plt.title('Total Sales by Product Category')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.show()
```

```python
# 4. Sales trend over time (Line Plot)
orders_df['OrderDate'] = pd.to_datetime(orders_df['OrderDate'])
sales_trend = orders_df.groupby(orders_df['OrderDate'].dt.to_period('M'))['TotalAmount'].sum().reset_index()
sales_trend['OrderDate'] = sales_trend['OrderDate'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
sns.lineplot(x='OrderDate', y='TotalAmount', data=sales_trend)
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 5. Correlation heatmap of numerical variables (Heatmap)
merged_df = orderdetails_df.merge(orders_df, on='OrderID').merge(products_df, on='ProductID')
numerical_columns = ['Quantity', 'UnitPrice', 'TotalPrice', 'TotalAmount', 'Price', 'StockQuantity']
correlation_matrix = merged_df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# 6. Most popular products
popular_products = orderdetails_df.groupby('ProductID')['Quantity'].sum().reset_index()
popular_products = popular_products.merge(products_df, on='ProductID').sort_values(by='Quantity', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='Quantity', y='ProductName', data=popular_products.head(10))
plt.title('Top 10 Most Popular Products')
plt.xlabel('Quantity Sold')
plt.ylabel('Product Name')
plt.show()
```

```python
# 7. Stock levels of products
plt.figure(figsize=(12, 6))
sns.histplot(products_df['StockQuantity'], bins=30, kde=True)
plt.title('Distribution of Stock Levels')
plt.xlabel('Stock Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Find the product with the highest stock
highest_stock_product = products_df.loc[products_df['StockQuantity'].idxmax()]
print("Product with the highest stock level:")
print(highest_stock_product)

# Find the product with the lowest stock
lowest_stock_product = products_df.loc[products_df['StockQuantity'].idxmin()]
print("\nProduct with the lowest stock level:")
print(lowest_stock_product)

# Distribution of Stock Levels
plt.figure(figsize=(12, 6))
sns.histplot(products_df['StockQuantity'], bins=30, kde=True)
plt.title('Distribution of Stock Levels')
plt.xlabel('Stock Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# 8. Supplier contribution to total sales
supplier_sales = product_sales.groupby('SupplierID')['TotalPrice'].sum().reset_index()
supplier_sales = supplier_sales.merge(suppliers_df, on='SupplierID').sort_values(by='TotalPrice', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalPrice', y='SupplierName', data=supplier_sales)
plt.title('Supplier Contribution to Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Supplier Name')
plt.show()
```

```python
# 9. Gender distribution of customers (Pie Chart)
gender_distribution = customers_df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
gender_distribution.plot.pie(autopct='%1.1f%%')
plt.title('Gender Distribution of Customers')
plt.ylabel('')
plt.show()
```

```python
# 10. Total sales by state
state_sales = orders_df.groupby('State')['TotalAmount'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalAmount', y='State', data=state_sales.sort_values(by='TotalAmount', ascending=False))
plt.title('Total Sales by State')
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.show()
```

```python
# 11. Group by 'City' and count the number of unique customers
customers_by_city = customers_df.groupby('City')['CustomerID'].nunique().reset_index()
customers_by_city.rename(columns={'CustomerID': 'CustomerCount'}, inplace=True)

# Sort the data by customer count for better visualization
customers_by_city = customers_by_city.sort_values(by='CustomerCount', ascending=False)

# Plot number of customers by city
plt.figure(figsize=(12, 6))
sns.barplot(x='CustomerCount', y='City', data=customers_by_city, hue='City', dodge=False, palette='viridis', legend=False)
plt.title('Number of Customers in Each City')
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.show()
```

## Step 7. Conclusion

Based on the data visualizations and analysis from the Departments, Products, Orders, OrderDetails, and Suppliers tables, we can derive several key insights about the online grocery store's operations and performance.

1. **Department and Product Analysis**
   - **Observation**: Certain departments, like fresh produce and dairy, have a higher number of products.
   - **Conclusion**: Focus on maintaining a wide variety of products in key departments to attract more customers and meet diverse needs.

2. **Order Trends and Patterns**
   - **Observation**: Peak order periods, such as weekends and holidays, were identified.
   - **Conclusion**: Prepare for peak periods by increasing inventory and staffing. Tailor marketing and inventory strategies to regional variations in order volumes.

3. **Order Details and Product Performance**
   - **Observation**: Analysis highlighted the most frequently ordered products and the distribution of unit prices.
   - **Conclusion**: Keep popular products well-stocked to avoid stockouts. Adjust pricing strategies to maximize revenue and plan promotions around best-selling products.

4. **Supplier Distribution and Relationships**
   - **Observation**: Major cities like Hyderabad, Bangalore, and Kolkata have the highest concentration of suppliers.
   - **Conclusion**: Strengthen supply chain networks in major cities and expand the supplier base in regions with fewer suppliers to improve service coverage and reduce delivery times.
